```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and the column is 'City'
city_counts = df['City'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 5))
plt.bar(city_counts.index, city_counts.values, color='maroon')
plt.xlabel("Cities")
plt.ylabel("Number of occurrences")
plt.title("Occurrences of Cities in the 'City' Column")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df' and the column is 'City'
city_counts = df['City'].value_counts()

# Create a list of unique colors (you can customize this)
colors = ['maroon', 'blue', 'green', 'purple', 'orange']

# Create the bar plot with unique colors
plt.figure(figsize=(10, 5))
plt.bar(city_counts.index, city_counts.values, color=colors)
plt.xlabel("Cities")
plt.ylabel("Number of occurrences")
plt.title("Occurrences of Cities in the 'City' Column")
plt.show()
```
